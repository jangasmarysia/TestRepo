**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

```python
pd.read_sql_query("PRAGMA table_info(main);", conn)
```

```python
QUERY = """
SELECT CompTotal
FROM main
WHERE CompTotal IS NOT NULL
LIMIT 20;
"""
pd.read_sql_query(QUERY, conn)
```

```python
pd.read_sql_query("""
SELECT 
  typeof(CompTotal) AS sqlite_type,
  CompTotal
FROM main
WHERE CompTotal IS NOT NULL
LIMIT 20;
""", conn)
```

### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
#pull in data column 
QUERY = """
SELECT CompTotal
FROM main
WHERE CompTotal IS NOT NULL;
"""
df_comp_total = pd.read_sql_query(QUERY, conn)
```

```python
#sanity check
df_comp_total.shape, df_comp_total.head()
```

```python
#Plot a histogram of CompTotal (Total Compensation).
upper = df_comp_total['CompTotal'].quantile(0.95)       #data was very hard to read, zoomed into percentile

plt.figure(figsize=(8, 5))
plt.hist(df_comp_total.loc[df_comp_total['CompTotal'] <= upper, 'CompTotal'], bins=50)
plt.xlabel('CompTotal (<= 95th percentile)')
plt.ylabel('Number of Respondents')
plt.title('Histogram of Total Compensation (Zoomed)')
plt.tight_layout()
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
#pull in age column
QUERY = """
SELECT Age
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say';
"""
df_age = pd.read_sql_query(QUERY, conn)

df_age.head(), df_age.shape
```

```python
#convert age to numeric cata
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_age['Age_numeric'] = df_age['Age'].map(age_mapping)
df_age = df_age.dropna(subset=['Age_numeric'])
```

```python
#create box plot
plt.figure(figsize=(6, 4))
plt.boxplot(df_age['Age_numeric'], showfliers=True)
plt.ylabel('Age (approximate years)')
plt.title('Box Plot of Age Distribution')
plt.tight_layout()
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
#pull in columns Age and Workexp
QUERY = """
SELECT Age, WorkExp
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND WorkExp IS NOT NULL;
"""
df_age_work = pd.read_sql_query(QUERY, conn)

df_age_work.head(), df_age_work.shape
```

```python
#convert age ranges 
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_age_work['Age_numeric'] = df_age_work['Age'].map(age_mapping)
```

```python
#clean workexp data ensure it is numeric
df_age_work['WorkExp'] = pd.to_numeric(
    df_age_work['WorkExp'], errors='coerce'
)

df_age_work = df_age_work.dropna(subset=['Age_numeric', 'WorkExp'])
```

```python
#create scatterplot
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
plt.scatter(
    df_age_work['Age_numeric'],
    df_age_work['WorkExp'],
    alpha=0.5
)

plt.xlabel('Age (approximate years)')
plt.ylabel('Work Experience (years)')
plt.title('Scatter Plot of Age vs Work Experience')
plt.tight_layout()
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
pd.read_sql_query("SELECT DISTINCT TimeSearching FROM main WHERE TimeSearching IS NOT NULL LIMIT 50;", conn)
```

```python
pd.read_sql_query("SELECT DISTINCT Frustration FROM main WHERE Frustration IS NOT NULL LIMIT 50;", conn)
```

```python
#Pull from SQLite
QUERY = """
SELECT Age, TimeSearching, Frustration
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND TimeSearching IS NOT NULL
  AND Frustration IS NOT NULL;
"""
df_bubble = pd.read_sql_query(QUERY, conn)
```

```python
#convert to numeric midpoints
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_bubble['Age_numeric'] = df_bubble['Age'].map(age_mapping)
```

```python
#convert to numeric midpoints
time_search_mapping = {
    'Less than 15 minutes a day': 0.25,
    '15-30 minutes a day': 0.375,
    '30-60 minutes a day': 0.75,
    '60-120 minutes a day': 1.5,
    'Over 120 minutes a day': 3
}

df_bubble['TimeSearching_num'] = df_bubble['TimeSearching'].map(time_search_mapping)
```

```python
#count frustation
df_bubble['Frustration_count'] = (
    df_bubble['Frustration']
    .str.split(';')
    .apply(len)
)
```

```python
df_bubble = df_bubble.dropna(
    subset=['Age_numeric', 'TimeSearching_num', 'Frustration_count']
)
df_bubble
```

```python
#create scatterplot

plt.figure(figsize=(8, 6))
plt.scatter(
    df_bubble['TimeSearching_num'],
    df_bubble['Frustration_count'],
    s=df_bubble['Age_numeric'] * 3,
    alpha=0.5
)

plt.xlabel('Time Spent Searching (hours per day)')
plt.ylabel('Number of Frustration Sources')
plt.title('Time Searching vs Frustration (Bubble Size = Age)')
plt.tight_layout()
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
#pull data
QUERY = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL;
"""
df_db = pd.read_sql_query(QUERY, conn)

df_db.head(), df_db.shape
```

```python
#split data and then count
db_counts = (
    df_db['DatabaseWantToWorkWith']
    .str.split(';')     # split multi-select answers
    .explode()          # one database per row
    .str.strip()        # clean whitespace
    .value_counts()     # count frequency
    .head(5)            # top 5
)

db_counts

```

```python
#create pie chart
explode = [0.1] + [0] * (len(db_counts) - 1)

plt.figure(figsize=(7, 7))
plt.pie(
    db_counts.values,
    labels=db_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode
)

plt.title('Top 5 Databases Respondents Want to Work With Next Year')
plt.tight_layout()
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
#pull columns
QUERY = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND TimeSearching IS NOT NULL
  AND TimeAnswering IS NOT NULL;
"""
df_time = pd.read_sql_query(QUERY, conn)
```

```python
#filter for age group
age_30_35 = ['25-34 years old', '35-44 years old']
df_time = df_time[df_time['Age'].isin(age_30_35)].copy()
```

```python
#Map numeric time search ranges
time_search_mapping = {
    'Less than 15 minutes a day': 0.25,
    '15-30 minutes a day': 0.375,
    '30-60 minutes a day': 0.75,
    '60-120 minutes a day': 1.5,
    'Over 120 minutes a day': 3
}
```

```python
#map numeric time answer mappings
time_answer_mapping = {
    'Less than 15 minutes a day': 0.25,
    '15-30 minutes a day': 0.375,
    '30-60 minutes a day': 0.75,
    '60-120 minutes a day': 1.5,
    'Over 120 minutes a day': 3
}
```

```python
# apply the mappings
df_time['TimeSearching_num'] = df_time['TimeSearching'].map(time_search_mapping)
df_time['TimeAnswering_num'] = df_time['TimeAnswering'].map(time_answer_mapping)

df_time = df_time.dropna(subset=['TimeSearching_num', 'TimeAnswering_num'])
```

```python
#calculate medians
median_searching = df_time['TimeSearching_num'].median()
median_answering = df_time['TimeAnswering_num'].median()

median_searching, median_answering
```

```python
#create stacked bar chart
plt.figure(figsize=(6, 5))

plt.bar(
    ['Age 30–35'],
    [median_searching],
    label='Time Searching'
)

plt.bar(
    ['Age 30–35'],
    [median_answering],
    bottom=[median_searching],
    label='Time Answering'
)

plt.ylabel('Median Time (hours per day)')
plt.title('Median Time Searching and Answering (Age 30–35)')
plt.legend()
plt.tight_layout()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## pull data
QUERY = """
SELECT Age, CompTotal
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND CompTotal IS NOT NULL;
"""
df_comp_age = pd.read_sql_query(QUERY, conn)
```

```python
#map age ranges
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_comp_age['Age_numeric'] = df_comp_age['Age'].map(age_mapping)
```

```python
#filter ages
df_comp_age = df_comp_age[
    df_comp_age['Age_numeric'].between(45, 60)
].copy()
```

```python
#calculate median comptotal by age
median_comp_by_age = (
    df_comp_age
    .groupby('Age_numeric')['CompTotal']
    .median()
    .reset_index()
)

median_comp_by_age
```

```python
#create line chart
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
plt.plot(
    median_comp_by_age['Age_numeric'],
    median_comp_by_age['CompTotal'],
    marker='o'
)

plt.xlabel('Age (approximate years)')
plt.ylabel('Median Total Compensation')
plt.title('Median Total Compensation by Age (45–60)')
plt.tight_layout()
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
#pull data
QUERY = """
SELECT MainBranch
FROM main
WHERE MainBranch IS NOT NULL;
"""
df_branch = pd.read_sql_query(QUERY, conn)

df_branch.head(), df_branch.shape
```
