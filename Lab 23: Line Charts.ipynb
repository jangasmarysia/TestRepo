### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
#Pull in data
QUERY = """
SELECT Age, ConvertedCompYearly
FROM survey
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND ConvertedCompYearly IS NOT NULL
"""
df_age = pd.read_sql_query(QUERY, conn)

df_age.head(), df_age.shape
```

```python
#convert compensation to numeric
df_age["ConvertedCompYearly"] = pd.to_numeric(
    df_age["ConvertedCompYearly"],
    errors="coerce"
)

df_age = df_age.dropna()
```

```python
#remove ouliers for compensation
Q1 = df_age["ConvertedCompYearly"].quantile(0.25)
Q3 = df_age["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_age = df_age[df_age["ConvertedCompYearly"] <= upper]

df_age.shape
```

```python
#aggregate median compensation by age
median_comp = (
    df_age
    .groupby("Age")["ConvertedCompYearly"]
    .median()
    .reset_index()
)

median_comp
```

```python
#order age groups
age_order = [
    "Under 18 years old",
    "18-24 years old",
    "25-34 years old",
    "35-44 years old",
    "45-54 years old",
    "55-64 years old",
    "65 years or older"
]

median_comp["Age"] = pd.Categorical(
    median_comp["Age"],
    categories=age_order,
    ordered=True
)

median_comp = median_comp.sort_values("Age")

median_comp
```

```python
#create line chart
plt.figure(figsize=(10, 6))

plt.plot(
    median_comp["Age"],
    median_comp["ConvertedCompYearly"],
    marker="o"
)

plt.xlabel("Age Group")
plt.ylabel("Median Yearly Compensation (USD)")
plt.title("Median Compensation by Age Group")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Pull the data
QUERY = """
SELECT WorkExp, ConvertedCompYearly
FROM survey
WHERE WorkExp IS NOT NULL
  AND ConvertedCompYearly IS NOT NULL
"""
df_exp = pd.read_sql_query(QUERY, conn)

df_exp.head(), df_exp.shape
```

```python
#convert experience and compensation to numeric
df_exp["WorkExp"] = pd.to_numeric(df_exp["WorkExp"], errors="coerce")
df_exp["ConvertedCompYearly"] = pd.to_numeric(
    df_exp["ConvertedCompYearly"], errors="coerce"
)

df_exp = df_exp.dropna()

df_exp.head()
```

```python
#remove outliers
Q1 = df_exp["ConvertedCompYearly"].quantile(0.25)
Q3 = df_exp["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_exp = df_exp[df_exp["ConvertedCompYearly"] <= upper]

df_exp.shape
```

```python
#Aggregate median compensation by experience
median_exp = (
    df_exp
    .groupby("WorkExp")["ConvertedCompYearly"]
    .median()
    .reset_index()
)

median_exp.head()
```

```python
#sort by experience
median_exp = median_exp.sort_values("WorkExp")
```

```python
#Create the line chart
plt.figure(figsize=(10, 6))

plt.plot(
    median_exp["WorkExp"],
    median_exp["ConvertedCompYearly"],
    marker="o"
)

plt.xlabel("Years of Work Experience")
plt.ylabel("Median Yearly Compensation (USD)")
plt.title("Median Compensation by Experience Level")
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels
