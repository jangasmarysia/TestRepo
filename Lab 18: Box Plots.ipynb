## Visualizing Data


### Task 1: Visualizing the Distribution of Data


**1. Box Plot of `CompTotal` (Total Compensation)**


Use a box plot to analyze the distribution and outliers in total compensation.


```python
#pull in data from CompTotal column 
QUERY = """
SELECT CompTotal
FROM main
WHERE CompTotal IS NOT NULL;
"""
df_comp_total = pd.read_sql_query(QUERY, conn)
```

```python
#use IQR to cleanup outliars
Q1 = df_comp_total['CompTotal'].quantile(0.25)
Q3 = df_comp_total['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

upper = Q3 + 1.5 * IQR

df_trimmed = df_comp_total[df_comp_total['CompTotal'] <= upper]
```

```python
#create box plot

plt.figure(figsize=(6, 4))
plt.boxplot(df_trimmed['CompTotal'], vert=False)
plt.xlabel('Total Compensation')
plt.title('Box Plot of Total Compensation')
plt.tight_layout()
plt.show()
```

**2. Box Plot of Age (converted to numeric values)**


Convert the `Age` column into numerical values and visualize the distribution.


```python
#pull data
QUERY = """
SELECT Age
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say';
"""
df_age = pd.read_sql_query(QUERY, conn)
```

```python
#map ages
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_age['Age_numeric'] = df_age['Age'].map(age_mapping)
df_age = df_age.dropna().copy()
```

```python
#create box plot
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.boxplot(df_age['Age_numeric'], showfliers=True)
plt.ylabel('Age (Numeric Approximation)')
plt.title('Box Plot of Age Distribution')
plt.tight_layout()
plt.show()
```

### Task 2: Visualizing Relationships in Data


**1. Box Plot of `CompTotal` Grouped by Age Groups:**


Visualize the distribution of compensation across different age groups.


```python
#pull in columns Age and CompTotal
QUERY = """
SELECT Age, CompTotal
FROM main
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND CompTotal IS NOT NULL;
"""
df_age_comp = pd.read_sql_query(QUERY, conn)

df_age_comp.head(), df_age_comp.shape
```

```python
#histogram was messy, too many outliars used IQR to clean up data
Q1 = df_age_comp['CompTotal'].quantile(0.25)
Q3 = df_age_comp['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

upper = Q3 + 1.5 * IQR

df_trimmed = df_age_comp[df_age_comp['CompTotal'] <= upper]
```

```python
#define age group 
age_groups = [
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
```

```python
#Data structure, loop once per age, pull comptotal for values in that age group
box_data = [
    df_trimmed.loc[df_trimmed['Age'] == age, 'CompTotal']
    for age in age_groups
]
```

```python
#create boxplot
plt.figure(figsize=(10, 6))

plt.boxplot(
    box_data,
    labels=age_groups,
    showfliers=True
)

plt.xlabel('Age Group')
plt.ylabel('Total Compensation')
plt.title('Distribution of Total Compensation by Age Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**2. Box Plot of `CompTotal` Grouped by Job Satisfaction (`JobSatPoints_6`):**


Examine how compensation varies based on job satisfaction levels.


```python
#pull data
QUERY = """
SELECT CompTotal, JobSatPoints_6
FROM main
WHERE CompTotal IS NOT NULL
  AND JobSatPoints_6 IS NOT NULL;
"""
df = pd.read_sql_query(QUERY, conn)
```

```python
#convert to numeric
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

df = df.dropna()
```

```python
#fix outliers in comptotal
Q1 = df['CompTotal'].quantile(0.25)
Q3 = df['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

upper = Q3 + 1.5 * IQR

df = df[df['CompTotal'] <= upper]
```

```python
#create bins
bins = [0, 33, 66, 100]
labels = ['Low Satisfaction', 'Medium Satisfaction', 'High Satisfaction']

df['JobSatBucket'] = pd.cut(
    df['JobSatPoints_6'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
#prep boxplot structure
box_data = [
    df.loc[df['JobSatBucket'] == label, 'CompTotal']
    for label in labels
]
```

```python
#create boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(
    box_data,
    tick_labels=labels,
    showfliers=True
)

plt.xlabel('Job Satisfaction Level')
plt.ylabel('Total Compensation')
plt.title('Total Compensation by Job Satisfaction Level')
plt.tight_layout()
plt.show()
```

### Task 3: Visualizing the Composition of Data


**1. Box Plot of `ConvertedCompYearly` for the Top 5 Developer Types:**


Analyze compensation across the top 5 developer roles.


```python
#pull data
QUERY = """
SELECT DevType, ConvertedCompYearly
FROM main
WHERE DevType IS NOT NULL
  AND ConvertedCompYearly IS NOT NULL;
"""
df_dev = pd.read_sql_query(QUERY, conn)
```

```python
#Split & Explode dev type data
df_dev['DevType'] = df_dev['DevType'].str.split(';')
df_dev = df_dev.explode('DevType')
```

```python
#find top 5 roles
top_dev_types = (
    df_dev['DevType']
    .value_counts()
    .head(5)
    .index
)
```

```python
#Filter to top 5 roles only
df_top5 = df_dev[df_dev['DevType'].isin(top_dev_types)].copy()
```

```python
#Convert compensation to numeric + remove extreme outliers (IQR)
df_top5['ConvertedCompYearly'] = pd.to_numeric(
    df_top5['ConvertedCompYearly'],
    errors='coerce'
)

df_top5 = df_top5.dropna()

Q1 = df_top5['ConvertedCompYearly'].quantile(0.25)
Q3 = df_top5['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_top5 = df_top5[df_top5['ConvertedCompYearly'] <= upper]
```

```python
#Prepare the boxplot
box_data = [
    df_top5.loc[df_top5['DevType'] == dev, 'ConvertedCompYearly']
    for dev in top_dev_types
]
```

```python
#create boxplot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot(
    box_data,
    tick_labels=top_dev_types,
    showfliers=True
)

plt.xticks(rotation=30, ha='right')
plt.ylabel('Yearly Compensation (USD)')
plt.title('ConvertedCompYearly by Top 5 Developer Roles')
plt.tight_layout()
plt.show()
```

**2. Box Plot of `CompTotal` for the Top 5 Countries:**


Analyze compensation across respondents from the top 5 countries.


```python
#pull data
QUERY = """
SELECT Country, CompTotal
FROM main
WHERE Country IS NOT NULL
  AND CompTotal IS NOT NULL;
"""
df_country = pd.read_sql_query(QUERY, conn)

df_country.head(), df_country.shape
```

```python
#find top 5 countries
top_countries = (
    df_country['Country']
    .value_counts()
    .head(5)
    .index
)
```

```python
#filter data
df_top5 = df_country[df_country['Country'].isin(top_countries)].copy()
```

```python
#convert comptotal to numeric

df_top5['CompTotal'] = pd.to_numeric(
    df_top5['CompTotal'],
    errors='coerce'
)

df_top5 = df_top5.dropna()
```

```python
#remove outliers
Q1 = df_top5['CompTotal'].quantile(0.25)
Q3 = df_top5['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

upper = Q3 + 1.5 * IQR

df_top5 = df_top5[df_top5['CompTotal'] <= upper]
```

```python
#prepare for the boxplot
box_data = [
    df_top5.loc[df_top5['Country'] == country, 'CompTotal']
    for country in top_countries
]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot(
    box_data,
    tick_labels=top_countries,
    showfliers=True
)

plt.xticks(rotation=30, ha='right')
plt.ylabel('Total Compensation')
plt.title('CompTotal Distribution for Top 5 Countries')
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data

Analyze compensation for different employment types.
