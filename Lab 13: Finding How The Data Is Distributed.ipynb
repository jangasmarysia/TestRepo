### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
#Prepare the language columns
# Keep only rows where at least one language column exists
df_langs = df[['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']].dropna(how='all').copy()

# Split into lists
df_langs['LanguageHaveWorkedWith'] = df_langs['LanguageHaveWorkedWith'].str.split(';')
df_langs['LanguageWantToWorkWith'] = df_langs['LanguageWantToWorkWith'].str.split(';')
```

```python
#Count Frequency for Have worked with
have_counts = (
    df_langs
    .explode('LanguageHaveWorkedWith')['LanguageHaveWorkedWith']
    .value_counts()
)
```

```python
#count frequence for want worked with
want_counts = (
    df_langs
    .explode('LanguageWantToWorkWith')['LanguageWantToWorkWith']
    .value_counts()
)
```

```python
#combine into one comparison table
language_compare = (
    pd.concat(
        [have_counts, want_counts],
        axis=1,
        keys=['HaveWorkedWith', 'WantToWorkWith']
    )
    .fillna(0)
    .astype(int)
)
```

```python
#only view top 10 languages
top_languages = (
    language_compare.sum(axis=1)
    .sort_values(ascending=False)
    .head(10)
    .index
)

language_compare = language_compare.loc[top_languages]
```

```python
#Grouped Bar chart

x = np.arange(len(language_compare))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, language_compare['HaveWorkedWith'], width, label='Have Worked With')
plt.bar(x + width/2, language_compare['WantToWorkWith'], width, label='Want to Work With')

plt.xticks(x, language_compare.index, rotation=45, ha='right')
plt.ylabel('Number of Respondents')
plt.title('Comparison of Programming Languages: Experience vs Interest')
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# Top 10 regions count
top_regions = (
    df['Country']
    .value_counts()
    .head(10)
    .index
)

df_region = df[df['Country'].isin(top_regions)].copy()
```

```python
#cross-tabulate RemoteWork by region
remote_region_pct = (
    pd.crosstab(
        df_region['Country'],
        df_region['RemoteWork'],
        normalize='index'
    )
    .mul(100)
    .round(2)
)

remote_region_pct
```

```python
#create heatmap

plt.figure(figsize=(10, 6))
plt.imshow(remote_region_pct, aspect='auto')
plt.colorbar(label='Percentage of Respondents')

plt.xticks(
    range(len(remote_region_pct.columns)),
    remote_region_pct.columns,
    rotation=45,
    ha='right'
)
plt.yticks(
    range(len(remote_region_pct.index)),
    remote_region_pct.index
)

plt.title('Remote Work Distribution by Region')
plt.xlabel('Remote Work Preference')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install scipy
```

```python
from scipy.stats import spearmanr
```

```python
#Prepare the data
df_corr = df[['JobSat', 'YearsCodePro']].copy()

# Ensure YearsCodePro is numeric
df_corr.loc[:, 'YearsCodePro'] = pd.to_numeric(
    df_corr['YearsCodePro'], errors='coerce'
)

# Drop missing values
df_corr = df_corr.dropna()
```

```python
#Calculate Spearman correlation
spearman_corr = df_corr['JobSat'].corr(
    df_corr['YearsCodePro'],
    method='spearman'
)

spearman_corr
```

```python
#calculate pearson
pearson_corr = df_corr['JobSat'].corr(
    df_corr['YearsCodePro'],
    method='pearson'
)

pearson_corr
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
#Only top 20 employment types
top_employment = (
    df_region['Employment']
    .value_counts()
    .head(20)
    .index
)
```

```python
#create df_top variable
df_top = df_region[df_region['Employment'].isin(top_employment)].copy()
```

```python
#cross-tabulate employment by region
employment_ed_pct = (
    pd.crosstab(
        df_top['Employment'],
        df_top['EdLevel'],
        normalize='index'
    )
    .mul(100)
    .round(2)
)

employment_ed_pct
```

```python
#create stacked bar chart
employment_ed_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

plt.xlabel('Education Level')
plt.ylabel('Percentage of Respondents')
plt.title('Employment Type Distribution by Education Level')
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
