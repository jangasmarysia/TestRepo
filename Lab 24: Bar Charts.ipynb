### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
#pull data
QUERY = """
SELECT ConvertedCompYearly
FROM survey
WHERE ConvertedCompYearly IS NOT NULL
"""
df_comp = pd.read_sql_query(QUERY, conn)

df_comp.head(), df_comp.shape
```

```python
#convert compensation to numeric
df_comp["ConvertedCompYearly"] = pd.to_numeric(
    df_comp["ConvertedCompYearly"],
    errors="coerce"
)

df_comp = df_comp.dropna()

df_comp.head()
```

```python
#remove outliers
Q1 = df_comp["ConvertedCompYearly"].quantile(0.25)
Q3 = df_comp["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_comp = df_comp[df_comp["ConvertedCompYearly"] <= upper]

df_comp.shape
```

```python
#create histogram
plt.figure(figsize=(10, 6))

plt.hist(
    df_comp["ConvertedCompYearly"],
    bins=30
)

plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Number of Respondents")
plt.title("Distribution of Yearly Compensation")

plt.tight_layout()
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
#pull data
QUERY = """
SELECT Age, ConvertedCompYearly
FROM survey
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND ConvertedCompYearly IS NOT NULL
"""
df_age = pd.read_sql_query(QUERY, conn)

df_age.head()
```

```python
#Convert compensation to numeric
df_age["ConvertedCompYearly"] = pd.to_numeric(
    df_age["ConvertedCompYearly"], errors="coerce"
)

df_age = df_age.dropna()
```

```python
#define age to number mapping
age_map = {
    "Under 18 years old": 1,
    "18-24 years old": 2,
    "25-34 years old": 3,
    "35-44 years old": 4,
    "45-54 years old": 5,
    "55-64 years old": 6,
    "65 years or older": 7
}
```

```python
#apply the mapping
df_age["Age_num"] = df_age["Age"].map(age_map)

df_age = df_age.dropna()

df_age[["Age", "Age_num"]].drop_duplicates().sort_values("Age_num")
```

```python
#create the box plot
plt.figure(figsize=(10,6))

plt.boxplot(
    [
        df_age.loc[df_age["Age_num"] == i, "ConvertedCompYearly"]
        for i in sorted(age_map.values())
    ],
    labels=age_map.keys()
)

plt.yscale("log")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Age Group")
plt.ylabel("Yearly Compensation")
plt.title("Compensation Distribution by Age Group")
plt.tight_layout()
plt.show()
```

### Task 2: Visualizing Relationships in Data

```python
#convert compensation to numeric
df_age["ConvertedCompYearly"] = pd.to_numeric(
    df_age["ConvertedCompYearly"], errors="coerce"
)

df_age = df_age.dropna()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
#pull data
QUERY = """
SELECT Age, ConvertedCompYearly, JobSatPoints_6
FROM survey
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND ConvertedCompYearly IS NOT NULL
  AND JobSatPoints_6 IS NOT NULL
"""
df_bubble = pd.read_sql_query(QUERY, conn)

df_bubble.head(), df_bubble.shape
```

```python
#convert to numeric
df_bubble["ConvertedCompYearly"] = pd.to_numeric(
    df_bubble["ConvertedCompYearly"], errors="coerce"
)
df_bubble["JobSatPoints_6"] = pd.to_numeric(
    df_bubble["JobSatPoints_6"], errors="coerce"
)

df_bubble = df_bubble.dropna()
```

```python
#map age to numeric
age_map = {
    "Under 18 years old": 1,
    "18-24 years old": 2,
    "25-34 years old": 3,
    "35-44 years old": 4,
    "45-54 years old": 5,
    "55-64 years old": 6,
    "65 years or older": 7
}

df_bubble["Age_numeric"] = df_bubble["Age"].map(age_map)
df_bubble = df_bubble.dropna()
```

```python
#remove outliers
Q1 = df_bubble["ConvertedCompYearly"].quantile(0.25)
Q3 = df_bubble["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_bubble = df_bubble[df_bubble["ConvertedCompYearly"] <= upper]
```

```python
#Create the bubble plot
plt.figure(figsize=(10,6))

scatter = plt.scatter(
    df_bubble["ConvertedCompYearly"],
    df_bubble["JobSatPoints_6"],
    c=df_bubble["Age_numeric"],
    cmap="viridis",
    alpha=0.5,
    s=40
)

cbar = plt.colorbar(scatter)
cbar.set_ticks(list(age_map.values()))
cbar.set_ticklabels(list(age_map.keys()))
cbar.set_label("Age Group")

plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Job Satisfaction (Points 6)")
plt.title("Compensation vs Job Satisfaction (Color = Age Group)")
plt.tight_layout()
plt.show()

```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
#pull data
QUERY = """
SELECT Age, JobSatPoints_6, JobSatPoints_7
FROM survey
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND JobSatPoints_6 IS NOT NULL
  AND JobSatPoints_7 IS NOT NULL
"""
df_sat = pd.read_sql_query(QUERY, conn)

df_sat.head(), df_sat.shape
```

```python
#Convert satisfaction columns to numeric
df_sat["JobSatPoints_6"] = pd.to_numeric(df_sat["JobSatPoints_6"], errors="coerce")
df_sat["JobSatPoints_7"] = pd.to_numeric(df_sat["JobSatPoints_7"], errors="coerce")

df_sat = df_sat.dropna()
```

```python
#Aggregate median satisfaciton by age
stacked = (
    df_sat
    .groupby("Age")
    .agg(
        sat6=("JobSatPoints_6", "median"),
        sat7=("JobSatPoints_7", "median")
    )
    .reset_index()
)

stacked
```

```python
#order by age group
age_order = [
    "Under 18 years old",
    "18-24 years old",
    "25-34 years old",
    "35-44 years old",
    "45-54 years old",
    "55-64 years old",
    "65 years or older"
]

stacked["Age"] = pd.Categorical(stacked["Age"], categories=age_order, ordered=True)
stacked = stacked.sort_values("Age")
```

```python
#create stacked bar chart
stacked.set_index("Age")[["sat6", "sat7"]].plot(
    kind="bar",
    stacked=True,
    figsize=(10,6)
)

plt.ylabel("Median Job Satisfaction")
plt.title("Median Job Satisfaction by Age Group")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
#pull the data
QUERY = """
SELECT Age, ConvertedCompYearly
FROM survey
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND ConvertedCompYearly IS NOT NULL
"""
df_age = pd.read_sql_query(QUERY, conn)

df_age.head(), df_age.shape
```

```python
#convert compensation to numeric
df_age["ConvertedCompYearly"] = pd.to_numeric(
    df_age["ConvertedCompYearly"], errors="coerce"
)

df_age = df_age.dropna()
```

```python
#Remove outliers
Q1 = df_age["ConvertedCompYearly"].quantile(0.25)
Q3 = df_age["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_age = df_age[df_age["ConvertedCompYearly"] <= upper]
```

```python
#compute median compensation by age group
median_comp = (
    df_age
    .groupby("Age")["ConvertedCompYearly"]
    .median()
    .reset_index()
)

median_comp
```

```python
#order age groups
age_order = [
    "Under 18 years old",
    "18-24 years old",
    "25-34 years old",
    "35-44 years old",
    "45-54 years old",
    "55-64 years old",
    "65 years or older"
]

median_comp["Age"] = pd.Categorical(
    median_comp["Age"],
    categories=age_order,
    ordered=True
)

median_comp = median_comp.sort_values("Age")
```

```python
#create the brouped bar chart
plt.figure(figsize=(10,6))

plt.bar(
    median_comp["Age"],
    median_comp["ConvertedCompYearly"]
)

plt.xlabel("Age Group")
plt.ylabel("Median Yearly Compensation (USD)")
plt.title("Median Compensation by Age Group")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
