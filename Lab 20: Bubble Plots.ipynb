### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
#Pull data
QUERY = """
SELECT Age, SOPartFreq, JobSat
FROM survey
WHERE Age IS NOT NULL
  AND SOPartFreq IS NOT NULL
  AND JobSat IS NOT NULL
  AND Age != 'Prefer not to say'
"""
df_bubble = pd.read_sql_query(QUERY, conn)

df_bubble.head(), df_bubble.shape
```

```python
age_map = {
    "Under 18 years old": 1,
    "18-24 years old": 2,
    "25-34 years old": 3,
    "35-44 years old": 4,
    "45-54 years old": 5,
    "55-64 years old": 6,
    "65 years or older": 7
}

freq_map = {
    "None": 1,
    "Less than once per month or monthly": 2,
    "A few times per week": 3,
    "Daily or almost daily": 4,
    "Multiple times per day": 5
}

df_bubble["Age_num"] = df_bubble["Age"].map(age_map)
df_bubble["Freq_num"] = df_bubble["SOPartFreq"].map(freq_map)

df_bubble = df_bubble.dropna()

df_bubble.head()
```

```python
#prepare data for the bubble plot
bubble_data = (
    df_bubble
    .groupby(["Age_num", "Freq_num"])
    .agg(
        avg_jobsat=("JobSat", "mean"),
        count=("JobSat", "size")
    )
    .reset_index()
)

bubble_data.head()
```

```python
#create bubble plot
plt.figure(figsize=(10, 6))

plt.scatter(
    bubble_data["Age_num"],
    bubble_data["Freq_num"],
    s=bubble_data["avg_jobsat"] * bubble_data["count"] * 2,
    alpha=0.5
)

plt.xlabel("Age Group (Young → Old)")
plt.ylabel("Participation Frequency (Low → High)")
plt.title("Age vs Participation Frequency (Bubble size = Job Satisfaction)")

plt.xticks(list(age_map.values()), list(age_map.keys()), rotation=45, ha="right")
plt.yticks(list(freq_map.values()), list(freq_map.keys()))

plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
#Pull data
QUERY = """
SELECT Age, JobSat, ConvertedCompYearly
FROM survey
WHERE ConvertedCompYearly IS NOT NULL
  AND JobSat IS NOT NULL
  AND Age IS NOT NULL
  AND Age != 'Prefer not to say'
"""
df_bubble2 = pd.read_sql_query(QUERY, conn)

df_bubble2.head(), df_bubble2.shape
```

```python
#convert age to numeric
age_map = {
    "Under 18 years old": 1,
    "18-24 years old": 2,
    "25-34 years old": 3,
    "35-44 years old": 4,
    "45-54 years old": 5,
    "55-64 years old": 6,
    "65 years or older": 7
}

df_bubble2["Age_num"] = df_bubble2["Age"].map(age_map)

df_bubble2 = df_bubble2.dropna()

df_bubble2.head()
```

```python
#convert compensation to numeric and remove outliers
df_bubble2["ConvertedCompYearly"] = pd.to_numeric(
    df_bubble2["ConvertedCompYearly"],
    errors="coerce"
)

df_bubble2 = df_bubble2.dropna()

Q1 = df_bubble2["ConvertedCompYearly"].quantile(0.25)
Q3 = df_bubble2["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_bubble2 = df_bubble2[df_bubble2["ConvertedCompYearly"] <= upper]

df_bubble2.shape
```

```python
df_sample = df_bubble2.sample(n=2000, random_state=42)

df_sample.shape
```

```python
#create the bubble plot
plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    df_sample["ConvertedCompYearly"],
    df_sample["JobSat"],
    c=df_sample["Age_num"],
    alpha=0.6
)

plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Job Satisfaction")
plt.title("Compensation vs Job Satisfaction (Color = Age group)")

cbar = plt.colorbar(scatter)
cbar.set_ticks(list(age_map.values()))
cbar.set_ticklabels(list(age_map.keys()))
cbar.set_label("Age Group")

plt.tight_layout()
plt.show()


```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
#Pull data
QUERY = """
SELECT Age, LanguageHaveWorkedWith
FROM survey
WHERE Age IS NOT NULL
  AND Age != 'Prefer not to say'
  AND LanguageHaveWorkedWith IS NOT NULL
"""
df_lang = pd.read_sql_query(QUERY, conn)

df_lang.head(), df_lang.shape
```

```python
#Split and explode the language column
df_lang["LanguageHaveWorkedWith"] = df_lang["LanguageHaveWorkedWith"].str.split(";")
df_lang = df_lang.explode("LanguageHaveWorkedWith")

df_lang["LanguageHaveWorkedWith"] = df_lang["LanguageHaveWorkedWith"].str.strip()

df_lang.head()
```

```python
#keep only the most common languages
top_languages = (
    df_lang["LanguageHaveWorkedWith"]
    .value_counts()
    .head(10)
    .index
)

df_lang = df_lang[df_lang["LanguageHaveWorkedWith"].isin(top_languages)]

df_lang.shape
```

```python
#convert age to numeric
age_map = {
    "Under 18 years old": 1,
    "18-24 years old": 2,
    "25-34 years old": 3,
    "35-44 years old": 4,
    "45-54 years old": 5,
    "55-64 years old": 6,
    "65 years or older": 7
}

df_lang["Age_num"] = df_lang["Age"].map(age_map)
df_lang = df_lang.dropna()

df_lang.head()
```

```python
#Aggregate counts for bubble sizes
bubble_lang = (
    df_lang
    .groupby(["LanguageHaveWorkedWith", "Age_num"])
    .size()
    .reset_index(name="count")
)

bubble_lang.head()
```

```python
#convert Launguages to numeric 
lang_codes = {lang: i+1 for i, lang in enumerate(top_languages)}
bubble_lang["Lang_num"] = bubble_lang["LanguageHaveWorkedWith"].map(lang_codes)
```

```python
#create bubble plot
plt.figure(figsize=(12, 7))

plt.scatter(
    bubble_lang["Lang_num"],
    bubble_lang["Age_num"],
    s=bubble_lang["count"] * 3,
    alpha=0.6
)

plt.xticks(list(lang_codes.values()), list(lang_codes.keys()), rotation=45, ha="right")
plt.yticks(list(age_map.values()), list(age_map.keys()))

plt.xlabel("Programming Language")
plt.ylabel("Age Group")
plt.title("Programming Language Popularity by Age Group (Bubble size = Frequency)")

plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
#Pull data
QUERY = """
SELECT DevType, ConvertedCompYearly, JobSat
FROM survey
WHERE DevType IS NOT NULL
  AND ConvertedCompYearly IS NOT NULL
  AND JobSat IS NOT NULL
"""
df_dev = pd.read_sql_query(QUERY, conn)

df_dev.head(), df_dev.shape
```

```python
# Split and exlode devtype
df_dev["DevType"] = df_dev["DevType"].str.split(";")
df_dev = df_dev.explode("DevType")

df_dev["DevType"] = df_dev["DevType"].str.strip()

df_dev.head()
```

```python
#convert compensation to numeric and remove extreme outliers
df_dev["ConvertedCompYearly"] = pd.to_numeric(
    df_dev["ConvertedCompYearly"],
    errors="coerce"
)

df_dev = df_dev.dropna()

Q1 = df_dev["ConvertedCompYearly"].quantile(0.25)
Q3 = df_dev["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_dev = df_dev[df_dev["ConvertedCompYearly"] <= upper]

df_dev.shape
```

```python
#keep only top developer roles
top_roles = (
    df_dev["DevType"]
    .value_counts()
    .head(10)
    .index
)

df_dev = df_dev[df_dev["DevType"].isin(top_roles)]

df_dev.shape
```

```python
#Aggregate compensation & job satisfaciton by role
bubble_dev = (
    df_dev
    .groupby("DevType")
    .agg(
        avg_comp=("ConvertedCompYearly", "mean"),
        avg_jobsat=("JobSat", "mean"),
        count=("JobSat", "size")
    )
    .reset_index()
)

bubble_dev
```

```python
#create bubble plot
plt.figure(figsize=(12, 7))

scatter = plt.scatter(
    bubble_dev["DevType"],
    bubble_dev["avg_comp"],
    s=bubble_dev["count"] * 2,
    c=bubble_dev["avg_jobsat"],
    alpha=0.7
)

plt.xticks(rotation=45, ha="right")
plt.xlabel("Developer Role")
plt.ylabel("Average Yearly Compensation (USD)")
plt.title("Compensation Across Developer Roles\n(Size = # of Respondents, Color = Job Satisfaction)")

cbar = plt.colorbar(scatter)
cbar.set_label("Average Job Satisfaction")

plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
#pull data
QUERY = """
SELECT Country, LanguageAdmired
FROM survey
WHERE Country IS NOT NULL
  AND LanguageAdmired IS NOT NULL
"""
df_admire = pd.read_sql_query(QUERY, conn)

df_admire.head(), df_admire.shape
```

```python
#Split and explode admired launguages
```

```python
df_admire["LanguageAdmired"] = df_admire["LanguageAdmired"].str.split(";")
df_admire = df_admire.explode("LanguageAdmired")

df_admire["LanguageAdmired"] = df_admire["LanguageAdmired"].str.strip()

df_admire.head()
```

```python
#keep only top admired languages and top countries
top_languages = (
    df_admire["LanguageAdmired"]
    .value_counts()
    .head(10)
    .index
)

top_countries = (
    df_admire["Country"]
    .value_counts()
    .head(10)
    .index
)

df_admire = df_admire[
    df_admire["LanguageAdmired"].isin(top_languages)
    & df_admire["Country"].isin(top_countries)
]

df_admire.shape
```

```python
#Aggregate admiration frequency
bubble_admire = (
    df_admire
    .groupby(["Country", "LanguageAdmired"])
    .size()
    .reset_index(name="count")
)

bubble_admire.head()
```

```python
#Encode countries and languages for pltotting
country_codes = {c: i+1 for i, c in enumerate(top_countries)}
lang_codes = {l: i+1 for i, l in enumerate(top_languages)}

bubble_admire["Country_num"] = bubble_admire["Country"].map(country_codes)
bubble_admire["Lang_num"] = bubble_admire["LanguageAdmired"].map(lang_codes)
```

```python
#create the bubble plot
plt.figure(figsize=(12, 8))

plt.scatter(
    bubble_admire["Country_num"],
    bubble_admire["Lang_num"],
    s=bubble_admire["count"] * 3,
    alpha=0.6
)

plt.xticks(list(country_codes.values()), list(country_codes.keys()), rotation=45, ha="right")
plt.yticks(list(lang_codes.values()), list(lang_codes.keys()))

plt.xlabel("Country")
plt.ylabel("Admired Programming Language")
plt.title("Admired Technologies Across Countries (Bubble size = Frequency)")

plt.tight_layout()
plt.show()
```
