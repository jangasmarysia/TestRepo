#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
import numpy as np
```

```python
df['YearsCodePro'].isnull().sum()
```

```python
df['YearsCodePro'].value_counts()
```

```python
#drop null values
df = df.dropna(subset=['YearsCodePro'])
```

```python
#drop "less than year" and "more then 50 data"
df = df.dropna(subset=['JobSat']).copy()

df.loc[:, 'YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df = df.dropna(subset=['YearsCodePro']).copy()
```

```python
#create experience ranges
bins = [0, 5, 10, 20, float('inf')]
labels = ['0–5', '5–10', '10–20', '20+']

df.loc[:, 'ExperienceRange'] = pd.cut(
    df['YearsCodePro'],
    bins=bins,
    labels=labels
)
```

```python
#Calculate the median JobSat for each range
median_jobsat = (
    df.groupby('ExperienceRange')['JobSat']
    .median()
    .reset_index()
)

median_jobsat
```

```python
#Visualize the relationship using a bar plot or similar visualization
import matplotlib.pyplot as plt

plt.figure()
plt.bar(median_jobsat['ExperienceRange'], median_jobsat['JobSat'])
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction')
plt.title('Median Job Satisfaction by Experience Level')
plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
# Count RemoteWork values
remote_counts = df['RemoteWork'].value_counts()

# Use a count plot to show remote work distribution.
plt.figure()
plt.bar(remote_counts.index, remote_counts.values)
plt.xlabel('Remote Work Preference')
plt.ylabel('Number of Respondents')
plt.title('Distribution of Remote Work Preferences')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.

role_remote_pct = (
    pd.crosstab(
        df['Employment'],
        df['RemoteWork'],
        normalize='index'
    )
    .mul(100)
    .round(2)
    .sort_values(by='Employment', ascending=False)
)

role_remote_pct

```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
df['LanguageHaveWorkedWith'].isnull().sum()
```

```python
df.shape
```

```python
df['Country'].isnull().sum()
```

```python
df['Country'].value_counts()
```

```python
#Use Top 10 countries 
top_countries = (
    df['Country']
    .value_counts()
    .head(10)
    .index
    .tolist()
)

top_countries
```

```python
#filter by top 10 countries
df_region = df[df['Country'].isin(top_countries)].copy()
```

```python
#language have worked with is a list, need to split it
df_region.loc[:, 'LanguageHaveWorkedWith'] = (
    df_region['LanguageHaveWorkedWith']
    .str.split(';')
)

df_lang = df_region.explode('LanguageHaveWorkedWith')
```

```python
#count languages by country
language_counts = (
    df_lang
    .groupby(['Country', 'LanguageHaveWorkedWith'])
    .size()
    .reset_index(name='Count')
)
```

```python
#keep top 10 languages
top_languages = (
    language_counts
    .sort_values(['Country', 'Count'], ascending=[True, False])
    .groupby('Country')
    .head(10)
)
```

```python
#build table for heatmap
heatmap_data = (
    df_lang
    .groupby(['Country', 'LanguageHaveWorkedWith'])
    .size()
    .reset_index(name='Count')
    .pivot(
        index='Country',
        columns='LanguageHaveWorkedWith',
        values='Count'
    )
    .fillna(0)
)
```

```python
#narrow down the amount of launguages
top_languages = (
    heatmap_data.sum()
    .sort_values(ascending=False)
    .head(15)
    .index
)

heatmap_data = heatmap_data[top_languages]
```

```python
#create heatmap table
plt.figure(figsize=(12, 6))
plt.imshow(heatmap_data, aspect='auto')
plt.colorbar(label='Number of Respondents')

plt.xticks(
    ticks=range(len(heatmap_data.columns)),
    labels=heatmap_data.columns,
    rotation=45,
    ha='right'
)

plt.yticks(
    ticks=range(len(heatmap_data.index)),
    labels=heatmap_data.index
)

plt.title('Programming Language Usage by Country')
plt.xlabel('Programming Language')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
#Prep the data
# Keep only rows with both values
df_scatter = df[['YearsCodePro', 'JobSatPoints_1']].copy()

# Convert to numeric just to be safe
df_scatter.loc[:, 'YearsCodePro'] = pd.to_numeric(
    df_scatter['YearsCodePro'], errors='coerce'
)

df_scatter.loc[:, 'JobSatPoints_1'] = pd.to_numeric(
    df_scatter['JobSatPoints_1'], errors='coerce'
)

# Drop rows with missing values
df_scatter = df_scatter.dropna()
```

```python
#create scatter plot
plt.figure()
plt.scatter(
    df_scatter['YearsCodePro'],
    df_scatter['JobSatPoints_1'],
    alpha=0.5
)

plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction Points')
plt.title('Relationship Between Experience and Job Satisfaction')
plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
#prep the data
df_edu_emp = df[['EdLevel', 'Employment']].dropna().copy()
```

```python
#cross tabulation
edu_emp_ct = pd.crosstab(
    df_edu_emp['EdLevel'],
    df_edu_emp['Employment']
)

edu_emp_ct
```

```python
#review by percentages
edu_emp_pct = (
    pd.crosstab(
        df_edu_emp['EdLevel'],
        df_edu_emp['Employment'],
        normalize='index'
    )
    .mul(100)
    .round(2)
)

edu_emp_pct
```

```python
#Stacked bar chart for visualization

import matplotlib.pyplot as plt

edu_emp_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 15)
)

plt.xlabel('Education Level')
plt.ylabel('Percentage of Respondents')
plt.title('Employment Type Distribution by Education Level')
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
