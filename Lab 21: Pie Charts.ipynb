### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
#pull data
QUERY = """
SELECT DatabaseWantToWorkWith
FROM survey
WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df_db = pd.read_sql_query(QUERY, conn)

df_db.head(), df_db.shape
```

```python
#split and explode the data
df_db["DatabaseWantToWorkWith"] = df_db["DatabaseWantToWorkWith"].str.split(";")
df_db = df_db.explode("DatabaseWantToWorkWith")

df_db["DatabaseWantToWorkWith"] = df_db["DatabaseWantToWorkWith"].str.strip()

df_db.head()
```

```python
#Count the top datbases
top5_dbs = (
    df_db["DatabaseWantToWorkWith"]
    .value_counts()
    .head(5)
)

top5_dbs
```

```python
#create the pie chart
plt.figure(figsize=(8, 8))

plt.pie(
    top5_dbs.values,
    labels=top5_dbs.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Top 5 Databases Respondents Want to Work With")
plt.show()
```
